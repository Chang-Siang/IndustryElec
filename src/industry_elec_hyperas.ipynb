```python
def MAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def RMSE(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

def calcError(origin, forecast):
    print("RMSE", RMSE(origin, forecast))
    print("MAPE", MAPE(origin, forecast))
```

```python
def buildTrain(train, pastDay=7, futureDay=7):
    X_train, Y_train = [], []
    for i in range(pastDay, len(train)-futureDay):
        X_train.append(train[i-pastDay:i])
        Y_train.append(train[i:i+futureDay, 0])
    X_train, Y_train = np.array(X_train), np.array(Y_train)  
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    return X_train, Y_train
```

```python
def buildTest(test, pastDay=7, futureDay=7):
    X_test = []
    for i in range(0, int(len(test)/futureDay)-(pastDay-futureDay)):
        X_test.append(test[futureDay*i:futureDay*i+pastDay])
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))  
    return X_test
```

```python
def forecast(model, input_data, shape):
    forecast_temp = model.predict(input_data)
    forecast = []
    for i in range(forecast_temp.shape[0]):
        forecast= np.concatenate((forecast, forecast_temp[i]), axis=0)
    forecast = np.reshape(forecast, (shape[0], shape[1]))
    return forecast
```

```python
def create_model(X, Y, Xv, Yv):
    epochs = 1000
    model = Sequential()
    model.add(SimpleRNN(units = {{choice([50,100,200,300,400])}}, activation='relu', input_shape = (X.shape[1], X.shape[2])))
    model.add(RepeatVector(7))
    model.add(SimpleRNN({{choice([50,100,200,300,400])}}, activation='relu', return_sequences=True))
    model.add(Dropout({{uniform(0, 0.3)}}))
    model.add(TimeDistributed(Dense({{choice([50,100,200,300,400])}}, activation='relu')))
    model.add(Dropout({{uniform(0, 0.3)}}))
    model.add(Flatten())
    model.add(Dense(units = 7))
    optimizer = optimizers.Adam(lr={{choice([0.0001,0.001,0.01,0.1])}}, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    model.compile(optimizer = optimizer, loss = 'mean_squared_error')
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor={{choice([0.01,0.1,0.3,0.5,0.7])}}, patience=20, mode='min', verbose=0)
    earlyStop=EarlyStopping(monitor="val_loss", patience=50, mode="min", restore_best_weights=True)
    model.fit(X, Y, 
              epochs=epochs, batch_size={{choice([16, 32, 64, 128])}}, 
              verbose=0, 
              validation_data=(Xv, Yv), 
              callbacks=[reduce_lr, earlyStop])
    return {'loss': min(model.history.history['val_loss']), 'status': STATUS_OK, 'model': model}
```
