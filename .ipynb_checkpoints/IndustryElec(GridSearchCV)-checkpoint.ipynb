### Sub：Metrics

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
def rmse(y_pred, y_true):
    return np.sqrt(((y_pred - y_true) ** 2).mean())
```

```python
def showPlot(true, prediction, title, size='M'):
    if(size=='L'):
        plt.figure(figsize=(24, 10))
    plt.plot(true, color = 'red', label = 'Real')
    plt.plot(prediction, color = 'blue', label = 'Prediction')
    plt.title('Industry Elec Prediction' + title + 't' + str(t))
    plt.xlabel('Time')
    plt.ylabel('kW')
    plt.legend()
    plt.show()
```

### Sub：Data

```python
features = ["工業用電(百萬度)", "isHoliday", "isHoliday(Next7)", "尖峰負載(MW)"]
# features_select = ["工業用電(百萬度)"]

def augFeatures(data, features_select=features):
    # 訓練資料擷取
    data = pd.DataFrame(data[features_select])
    data.shape
    return data
```

```python
from sklearn.preprocessing import MinMaxScaler
def normalize(train):
    # Feature Scaling
    sc = MinMaxScaler(feature_range = (0, 1))
    training_set_scaled = sc.fit_transform(training_set)
    return sc, training_set_scaled 
```

```python
def buildTrain(train, pastDay=7, futureDay=7):
    X_train, Y_train = [], []
    for i in range(pastDay, len(train)-futureDay):
        X_train.append(train[i-pastDay:i])
        Y_train.append(train[i:i+futureDay, 0])

    # 轉成numpy array的格式，以利輸入 RNN    
    X_train, Y_train = np.array(X_train), np.array(Y_train)  

    # Reshape
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    
    return X_train, Y_train
```

```python
def buildTest(test_set, training_set, sc, pastDay, futureDay, pre_step):
    dataset_total = augFeatures(df)
    dataset_total = dataset_total[:len(training_set) + (pre_step*futureDay)]

    inputs = dataset_total[len(dataset_total) - len(test_set) - pastDay:].values
    inputs = inputs.reshape(-1, dataset_total.shape[1])
    # Feature Scaling
    inputs = sc.transform(inputs)

    X_test = []
    for i in range(0, pre_step):
        X_test.append(inputs[futureDay*i:futureDay*i + (pastDay)])
    X_test = np.array(X_test)
    
    # Reshape
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))  
    
    return X_test
```

```python
def splitData(X, Y, rate):
    X_train = X[int(X.shape[0]*rate):]
    Y_train = Y[int(Y.shape[0]*rate):]
    X_val = X[:int(X.shape[0]*rate)]
    Y_val = Y[:int(Y.shape[0]*rate)]
    return X_train, Y_train, X_val, Y_val
```

### Sub：Model

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
```

```python
# Import the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import GRU
from keras.layers import CuDNNLSTM
from keras.layers import SimpleRNN
from keras.layers import Dropout
from keras.layers import BatchNormalization, Bidirectional
from keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
def create_model(shape=[30, 2], dropout_rate=0.0, learn_rate=0.0001, neurons=50):
    model = Sequential()
    model.add(CuDNNLSTM(units = neurons, return_sequences = True, input_shape = (shape[1], shape[2])))
    model.add(Dropout(dropout_rate))

    model.add(CuDNNLSTM(units = neurons, return_sequences = True))
    model.add(Dropout(dropout_rate))

    model.add(CuDNNLSTM(units = neurons))
    model.add(Dropout(dropout_rate))

    model.add(Dense(units = 7))
    model.compile(loss='mean_squared_error', optimizer=RMSprop(lr=learn_rate), metrics=['mape'])
#     model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model
```

## 訓練
- 使用 2016年1月至2018年12月的資料進行訓練
- 使用 2019年1月至2019年9月的資料進行測試

```python
RawTrain = df[0:1096]
# RawTrain = df[366:1096]

RawTest = df[1096:]
```

```python
# 特徵選取
training_set = augFeatures(RawTrain)

# 資料前處理
sc, training_set_scaled = normalize(training_set)

# 自定義變數
Timesteps = 7
OutputDay = 7

# 資料打包
X_train, Y_train = buildTrain(training_set_scaled, Timesteps, OutputDay)

# 驗證資料
# X, Y, Xv, Yv= splitData(X_train, Y_train, 0.1)
```

---
## GridSearchCV

```python
# fix random seed for reproducibility
model = KerasRegressor(build_fn=create_model, verbose=1, shape=X_train.shape)

# define the grid search parameters
batch_size = [16, 32, 64]
epochs = [100, 200, 400]
dropout_rate = [0.0, 0.2, 0.3, 0.5]
learn_rate = [0.1, 0.001, 0.0001, 0.00001]
neurons = [50, 100, 200]
param_grid = dict(batch_size=batch_size, 
                  epochs=epochs, 
                  dropout_rate=dropout_rate, 
                  learn_rate=learn_rate, 
                  neurons=neurons)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1, cv=3, scoring='neg_mean_absolute_error')
grid_result = grid.fit(X_train, Y_train)
```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
print(grid_result.cv_results_)
```

## Test-Prediction

```python
# 自定義變數
# Pre_step = 1
# Pre_step = 4
Pre_step = 39
```

```python
# 真實值
pre_time = 7 * Pre_step
test_set = pd.DataFrame(RawTest[["工業用電(百萬度)"]])
test_set = test_set[:pre_time]
real_elec = test_set.values

# 訓練資料
X_test = buildTest(test_set ,training_set, sc, Timesteps, OutputDay, Pre_step)
```

```python
test_predict = model.predict(X_test)
# test_predict = grid.predict(X_test)

test_predict = test_predict * (sc.data_max_[0]-sc.data_min_[0]) + sc.data_min_[0]

# Append Data
test_prediction = []
for i in range(Pre_step):
    test_prediction= np.concatenate((test_prediction, test_predict[i]), axis=0)
```

```python
# RMSE
print("RMSE: ", rmse(real_elec, test_prediction))
# MAPE
print("mape: ", mean_absolute_percentage_error(real_elec, test_prediction))
# plot
showPlot(real_elec, test_prediction, '(cLSTM)' + str(pre_time))
```
