```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

---
### Sub：Metrics

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
def rmse(y_pred, y_true):
    return np.sqrt(((y_pred - y_true) ** 2).mean())
```

```python
def showPlot(true, prediction, title, size='M'):
    if(size=='L'):
        plt.figure(figsize=(28, 10))
    plt.plot(true, color = 'red', label = 'Real')
    plt.plot(prediction, color = 'blue', label = 'Prediction')
    plt.title('Industry Elec Prediction' + title)
    plt.xlabel('Time')
    plt.ylabel('kW')
    plt.legend()
    plt.show()
```

### Sub：Data

```python
test_prediction = pd.read_csv('mean_test.csv')
test_prediction = pd.DataFrame(test_prediction[['kw']]).values
real_elec = pd.read_csv('test_trurh.csv')
real_elec = pd.DataFrame(real_elec[['kw']]).values

print('test_prediction', test_prediction.shape)
print('real_elec', real_elec.shape)
```

```python
print("RMSE: ", rmse(real_elec, test_prediction))
print("mape: ", mean_absolute_percentage_error(real_elec, test_prediction))
showPlot(real_elec, test_prediction, '(cLSTM)')
```
